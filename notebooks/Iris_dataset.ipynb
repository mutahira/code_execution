```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create results directory if it doesn't exist
if not os.path.exists('/content/results'):
    os.makedirs('/content/results')

# Load the Iris dataset
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                   header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.savefig('/content/results/scatter_plot_iris_sepal.png')
plt.show()
```
