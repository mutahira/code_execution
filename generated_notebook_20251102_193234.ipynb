```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Loading dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/flights.csv"
data = pd.read_csv(url)

# Step 2: Calculating total passengers per year
data['year'] = data['year'].astype(int)
passengers_per_year = data.groupby('year')['passengers'].sum().reset_index()

# Step 3: Plotting the trend
plt.figure(figsize=(10, 5))
plt.plot(passengers_per_year['year'], passengers_per_year['passengers'], marker='o')
plt.title('Trend of Number of Passengers per Year')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.grid()
plt.xticks(passengers_per_year['year'], rotation=45)
plt.tight_layout()
plt.show()
```
