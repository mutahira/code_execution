```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Loading dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/titanic.csv"
titanic_data = pd.read_csv(url)

# Step 2: Creating a count plot of survivors based on class and sex
plt.figure(figsize=(12, 6))
sns.countplot(data=titanic_data, x='class', hue='sex', palette='pastel', 
               order=['First', 'Second', 'Third'])
plt.title('Survivors based on Class and Sex')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Sex')
plt.show()
```
