```
import pandas as pd
import matplotlib.pyplot as plt
import os

# Step 1: Loading dataset
data_path = '/content/q8mszviz.csv'
data = pd.read_csv(data_path)

# Step 2: Calculating frequency distribution of foraging trip duration days
frequency_distribution = data['duration_round'].value_counts().sort_index()

# Step 3: Plotting the bar plot
plt.figure(figsize=(10, 5))
frequency_distribution.plot(kind='bar', color='skyblue')
plt.title('Frequency Distribution of Foraging Trip Duration Days')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Step 4: Saving the plot
results_dir = '/content/results'
os.makedirs(results_dir, exist_ok=True)
plt.savefig(os.path.join(results_dir, 'foraging_trip_duration_distribution.png'))

plt.show()
```
